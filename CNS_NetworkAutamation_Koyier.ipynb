### Define device credentials

We use a **Cisco DevNet Always-On Sandbox**. This is a real router that is always reachable over the internet

"Connect Handler" acts as a universal adapter

#### Roles of Connect Handler
-Establishes SSH connection to network devices

-Automatically handles "hanshake" (logging in, handles the '#' or '>' prompts)

**NB:** The momment you specify a device_type e.g. cisco_ios, `ConnectHandler` knows how to:

-specifically navigate the specific OS,

-enter navigation mode

-wait for the device to finish processing a command



```python
from netmiko import ConnectHandler # the primary function within netmiko. Different vendors use different CLIs
import getpass

# Device details for Cisco DevNet Sandbox

device = {
'device_type': 'cisco_xe',
'host': 'sandbox-iosxe-latest-1.cisco.com',
'username': 'developer',
'password': 'Ry@n123', #Public since we aint using getpass
'port': 22,
}

print(f"Targeting: {device['host']}")
```

### Retrive information (Show the commands)

Goals:

- Establish a connection and

- Pull the interface status

This is the "Read" part of the automation

```python
try:
  #Establish ssh connection
  print("Connecting to the device ...")
  net_connect = ConnectHandler(**device)

  #Run a show command
  output = net_connect.send_command('show ip int brief')
  print("\n-- Interface Status ---")
  print(output)

  #Disconnect
  net_connect.disconnect()

except Exception as e:
    print(f"Error connecting to the device: {e}")
```

## Pushing Configurarions

Automation is particularly spectacular when you eant to change settings

Next, let us create a new virtual interface called `loopback`

```python
## Re-establish connection for the config change

net_connect = ConnectHandler(**device)

config_commands = [
    'Interface Loopback99'
    'description Configured via python Netmiko'
    'ip address 1.1.1.1 255.255.255.255'
]

print ("Sending configuration commands...")
output = net_connect.send_config_set(config_commands)
print(output)

# Verify changes
verification = net_connect.send_command("show ip int brief | include loopback99")
print("\nVerification")
print(verification)

net_connect.disconnect()
```
